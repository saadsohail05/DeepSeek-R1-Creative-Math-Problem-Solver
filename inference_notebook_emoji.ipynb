```python
from fastapi import FastAPI
from pydantic import BaseModel
from transformers import AutoTokenizer
import torch
from pyngrok import ngrok
import nest_asyncio
import uvicorn
from unsloth import FastLanguageModel  # Import Unsloth

# Apply nest_asyncio to avoid event loop issues
nest_asyncio.apply()

# Define model name
model_name = "saadsohail/DeepSeekR1-GRPO-CreativeMathProblemSolver"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model with default parameters
model, tokenizer = FastLanguageModel.from_pretrained(model_name)
```

```python
# Load and prep dataset
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
...
</reasoning>
<answer>
...
</answer>
"""

XML_COT_FORMAT = """\
<reasoning>
{reasoning}
</reasoning>
<answer>
{answer}
</answer>
"""
```

```python
from fastapi.responses import JSONResponse
# Detect device
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Initialize FastAPI app
app = FastAPI()

# Request body schema
class EmojiRequest(BaseModel):
    emoji: str

@app.post("/generate")
async def generate_emoji_soluton(request: EmojiRequest):
    text = tokenizer.apply_chat_template([
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": request.emoji},
    ], tokenize=False, add_generation_prompt=True)

    input_ids = tokenizer(text, return_tensors="pt").input_ids.to(device)

    output_ids = model.generate(
        input_ids,
        max_new_tokens=1024,
        temperature=0.8,
        top_p=0.95
    )

    # Decode model output
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Format response into XML format
    formatted_response = XML_COT_FORMAT.format(reasoning="Generated reasoning here", answer=output_text)

    print(formatted_response)
    return JSONResponse(content={"solution": formatted_response})

# Start ngrok tunnel
public_url = ngrok.connect(7860).public_url
print(f"Public API URL: {public_url}")

uvicorn.run(app, host="0.0.0.0", port=7860)
```
